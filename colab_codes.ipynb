```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange (0,10)
y = np.arange (11,21)
```

```python
plt.scatter (x,y,c = 'b')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("scatter plot")
plt.show()
```
