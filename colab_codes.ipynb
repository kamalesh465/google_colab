```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.arange (0,10)
y = np.arange (11,21)
```

```python
plt.scatter (x,y,c = 'b')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("scatter plot")
plt.show()
```

```python
####creating subplots
plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'g')
plt.subplot(2,2,3)
plt.plot(x,y,'*--')
```

```python
a = np.array([22,87,5,43,56,73,55,54,27])
plt.hist(a)
plt.title("histo graph")
```

```python
x = np.arange(0,4*np.pi,0.1)
y = np.sin(x)
plt.title(" sine wave")
plt.plot(x,y)
```
