```
labels = 'python','c++','ruby','java'
size = [215,130,245,210]
colours = ['gold','yellowgreen','lightcoral','lightskyblue']
explode = (0.1,0,0,0)# explode ist slice
#plot
plt.pie(size,explode = explode,labels = labels ,colors = colours,autopct='%1.1f%%',shadow = True)
plt.axis('equal')
plt.show()
```
